### Automating the process with template matching

```python
template_= img[300:900,500:1000]
```

```python
plt.imshow(template_)
```

```python
result=cv2.matchTemplate(img, template_,cv2.TM_CCOEFF_NORMED)
```

```python
result.shape
```

```python
plt.imshow(result,cmap="gray")
```

```python
cv2.minMaxLoc(result)
```

```python
left=cv2.minMaxLoc(result)[3]
```

```python
left
```

```python
l,w,h= template_.shape
```

```python
right=(left[0]+l,left[1]+w)
```
